## **Code Section**

```python
import numpy
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv("data.csv")
```

```python
data.head()
```

```python
week1=data.iloc[0:,2].values
week2=data.iloc[0:,3].values
week3=data.iloc[0:,5].values
```

```python
week1
```

```python
company= data.iloc[:,0].values
```

```python
change = data.iloc[:,6].values
```

```python
clr=[]
if change[0]<0:
    clr="red"
else:
    clr="#2ecc71"
```

```python
clr
```

Bar Plot

```python
plt.figure(figsize=(25,15))
plt.xlabel("Name of the Companies")
plt.ylabel("Percentage in their Shares")
plt.title("4 August Nifty 50 (Top 20 Stocks)")
plt.bar(company,change,color=clr )
```

```python
plt.figure(figsize=(25,15))
plt.xlabel("Name of the Companies")
plt.ylabel("Percentage in their Shares")
plt.title("4 August Nifty 50 (Top 20 Stocks)")
plt.bar(company,change,color=clr )
```

Multi Volum Chart


```python
plt.figure(figsize=(35,25))
plt.xlabel("Name of the Companies")
plt.ylabel("Percentage in their Shares")
plt.title("4 August Nifty 50 (Top 20 Stocks)")
plt.bar(company,week1,color="red" )
plt.bar(company,week2,color="blue" )
plt.bar(company,week3,color="green" )
plt.legend()
plt.show()
```
